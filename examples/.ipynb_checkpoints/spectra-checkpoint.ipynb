# Acceleration response spectra

This notebook shows the computation of the accelartion response spectra of a time series.

## Details

```python
%matplotlib notebook
from __future__ import division
import matplotlib.pyplot as plt
import signals as sig
import lectura as lec
from IPython.display import Image
```

```python
ndats, signal = lec.readsignal_ncol('Gilroy.out' , 5 , 0.1)
lec.grafsignalG(signal , 'Gilroy' , 'Acceleration' , '(g)' , -0.1 , 0.1 , 0.005 , 0)
Sa , Ta = sig.respesp(signal, ndats, 0.005, 5.0, 50)
plt.figure(1)
plt.grid()
plt.xlabel('Period (s)')
plt.ylabel('Sa (g)')
plt.xlim(0.001,5.0)
plt.plot(Ta, Sa)
```
